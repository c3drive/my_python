### .merge()

```python
df1.merge(df2)
```

```python
df1 = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2'],})
df2 = pd.DataFrame({'key': ['k0', 'k1', 'kc'],
                    'C': ['c0', 'c1', 'c3'],
                    'D': ['d0', 'd1', 'd3'],})
```

```python
#　それぞれのデータフレームを結合できるところだけを残す
df1.merge(df2, how='inner', on='key')
```

```python
#　それぞれのデータフレームを結合できないものも全て表示する
df1.merge(df2, how='outer', on='key')
```

```python
#　それぞれのデータフレームの左側だけ結合できないものも全て表示する
df1.merge(df2, how='left', on='key')
```

```python
df1 = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                    'ID': ['aa', 'bb', 'cc'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2'],})
df2 = pd.DataFrame({'key': ['k0', 'k1', 'k2'],
                    'ID': ['aa', 'bb', 'cc'],
                    'C': ['c0', 'c1', 'c2'],
                    'D': ['d0', 'd1', 'd2'],})
```

```python
df1.merge(df2, on='key')
```

```python
df1.merge(df2, on='key', suffixes=('_right', '_left'))
```

```python
df1 = pd.DataFrame({'key1': ['k0', 'k1', 'k2'],
                    'A': ['a0', 'a1', 'a2'],
                    'B': ['b0', 'b1', 'b2'],})
df2 = pd.DataFrame({'key2': ['k0', 'k1', 'k2'],
                    'C': ['c0', 'c1', 'c2'],
                    'D': ['d0', 'd1', 'd2'],})
```

```python
# 左側でキーにしたいカラム、右側でキーにしたいカラムを指定する
df1.merge(df2, left_on='key1', right_on='key2')
```

## DataDrameのIteration
### apply()
関数を各行に適用(apply）する

```python
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
# revenue - budget
df.head(1)
```

```python
# 基本的な使い方。
# 例えば以下のような関数を作成する
# return_nan_if_zero(0) -- > nan
# return_nan_if_zero(10000) --> 10000
def return_nan_if_zero(budget):
    if budget == 0:
        return np.nan
    else:
        return budget
return_nan_if_zero(0)
```

```python
df['budget'].apply(return_nan_if_zero)
```

```python
df['budget'] = df['budget'].apply(return_nan_if_zero)
```

```python
df['budget']
```

```python
# Lambda関数を使うのが一般的
# まずはif文を1行に書いてみる
budget = 0
np.nan if budget == 0 else budget
```

```python
# Lambda関数にしてみる
x = lambda budget: np.nan if budget == 0 else budget
```

```python
x(budget)
```

```python
# 使ってみる
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
df['budget'] = df['budget'].apply(lambda x: np.nan if x == 0 else x)
df['revenue'] = df['revenue'].apply(lambda x: np.nan if x == 0 else x)
```

```python
df['budget']
```

```python
df['revenue']
```

```python
# budgetとrevenueがNaNのものをドロップする
df = df.dropna(subset=['budget', 'revenue'])
df
```

```python
# profit = revenue - budget
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
```

```python
df
```

```python
# profitが最大のものを取得したい（maxの値ではなくprofitがmaxのデータを取得
df.iloc[df['profit'].idxmax()]
```

```python
# こちらの方がシンプル
df[df['profit'] == df['profit'].max()]
```

## Pivot tableの作成
### .pivot_table()
データ処理の時にはあまり使わない。誰かに見せるときに使う

```python
# ユーザの支払いトランザクションのテーブル
data = {
    'Date': ['Jan-1', 'Jan-1', 'Jan-1', 'Jan-2'],
    'User': ['Emily', 'John', 'Nick', 'Emily'],
    'Method': ['Card', 'Card', 'Cash', 'Cash'],
    'Price': [100, 100, 300, 200]
}
df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot_table(values='Price', index=['Date', 'User'], columns=['Method'])
pivot_df
```

```python
pivot_df = df.pivot_table(values='Price', index=['Date', 'Method'], columns=['User'])
pivot_df
```

```python
# クロスセクション（それぞれの行を跨いでとってくる）
pivot_df.xs('Cash', level='Method')
```

## DataDrameの保存
### .to_csv()

```python
# profit(revenue - budget)の高い順に並べた表を作成してみる（０のものを除く）
```

```python
# 1. csvの読み込み
df = pd.read_csv('dataset/tmdb_5000_movies.csv')
```

```python
# 2. budgetやrevenueが0のものをfilter out
df = df[~(df['budget'] == 0) | (df['revenue'] == 0)]
```

```python
# 3. profitというカラムを作成
df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)
```

```python
# 4. profitでsortする
df = df.sort_values('profit', ascending=False)
```

```python
# 5.　保存
df.to_csv('dataset/tmdb_5000_movies_profit_sorted.csv')
```
